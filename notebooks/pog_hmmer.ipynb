### POG and SCG protein hit count for all samples and cutoffs

Create the required dataframes for all samples and cutoffs

```python
# uses pip install git+git://github.com/jakevdp/ipywidgets.git@a7a958b6e8c970361d25d9705236d1943470eed2
from ipywidgets import StaticInteract, RangeWidget, RadioWidget

def plot_pog_scg_count(sample, cutoff):
    # construct dfs
    hmmvq = pd.read_csv(FILE_REQS["POG_ALLVQ_FILES"].format(sample=sample),
                    delim_whitespace=True,
                    comment="#",
                    names=["target", "acc_t", "query", "acc_q",
                           "full_eval", "full_score", "full_bias", # full sequence
                           "dom1_eval", "dom1_score", "dom1_bias", # best 1 domain
                           "dom_exp", "dom_reg", "dom_clu", "dom_ov", # domain nr estimation
                           "dom_env", "dom_dom", "dom_rep", "dom_inc", "dom_desc_t", # cont'd: domain nr estimation
                           ])
    hmmvq["contig"] = hmmvq["query"].apply(lambda x: x.split("_")[0])
    clustdf = pd.read_csv(FILE_REQS["CLUSTER_FILES"].format(sample=sample, cutoff=cutoff),
                          names=["contig","cluster_gt{c}".format(c=cutoff)])
    hmmvq_c = hmmvq.merge(clustdf, how="inner", on="contig")
    hmmvq_best = hmmvq_c[hmmvq_c.groupby("query").dom1_score.transform(max) == hmmvq_c.dom1_score]
    cdd_rps = pd.read_csv(FILE_REQS["CDD_RPS_FILES"].format(sample=sample),
                      delim_whitespace=True,
                       names=["qseqid", "sseqid" , "evalue", "pident", "score", "qstart",
                              "qend", "sstart", "send", "length", "slen"])
    cdd_rps["COG"] = cdd_rps.sseqid.apply(lambda x: cdd2cog[x.split("|")[2]])
    cdd_rps["contig"] = cdd_rps.qseqid.apply(lambda x: x.split("_")[0])
    cdd_rps_c = cdd_rps.merge(clustdf, how="inner", on="contig")
    scg_rps = cdd_rps_c[cdd_rps_c.COG.isin(scgs_cogs)]
    hmmvq_best_pog = hmmvq_best.rename(columns={"target":"POG"}).merge(pog_vq,on="POG")
    
    # plot it
    fig, ax = plt.subplots(figsize=(16, 6),
                           subplot_kw={'axisbg':'#EEEEEE',
                                       'axisbelow':True})
    df1 = pd.DataFrame(scg_rps.groupby("cluster_gt{c}".format(c=cutoff)).COG.count())
    df2 = pd.DataFrame(hmmvq_best_pog[hmmvq_best_pog["Viral Quotient"] == 1].groupby("cluster_gt{c}".format(c=cutoff)).POG.count())
    pd.concat([df1.rename(columns={"COG":"SCG"}), df2], axis=1).plot(kind="bar",
                                                                     ax=ax,
                                                                     title="SCG and POG counts per cluster_gt{c} {sample}".format(c=cutoff,
                                                                                                                                  sample=sample))
    ax.set_ylim(0, 200)
    return fig
```

```python
StaticInteract(plot_pog_scg_count,
               cutoff=RadioWidget(PARAMETERS["cutoff"]),
               sample=RadioWidget(PARAMETERS["samples"]))
```

### Bin sizes

```python
# uses pip install git+git://github.com/jakevdp/ipywidgets.git@a7a958b6e8c970361d25d9705236d1943470eed2
from ipywidgets import StaticInteract, RangeWidget, RadioWidget
# pip install biopython
from Bio import SeqIO

def plot_bin_overview(sample, cutoff):
    # construct dfs
    hmmvq = pd.read_csv(FILE_REQS["POG_ALLVQ_FILES"].format(sample=sample),
                    delim_whitespace=True,
                    comment="#",
                    names=["target", "acc_t", "query", "acc_q",
                           "full_eval", "full_score", "full_bias", # full sequence
                           "dom1_eval", "dom1_score", "dom1_bias", # best 1 domain
                           "dom_exp", "dom_reg", "dom_clu", "dom_ov", # domain nr estimation
                           "dom_env", "dom_dom", "dom_rep", "dom_inc", "dom_desc_t", # cont'd: domain nr estimation
                           ])
    hmmvq["contig"] = hmmvq["query"].apply(lambda x: x.split("_")[0])
    clustdf = pd.read_csv(FILE_REQS["CLUSTER_FILES"].format(sample=sample, cutoff=cutoff),
                          names=["contig","cluster_gt{c}".format(c=cutoff)])
    hmmvq_c = hmmvq.merge(clustdf, how="inner", on="contig")
    hmmvq_best = hmmvq_c[hmmvq_c.groupby("query").dom1_score.transform(max) == hmmvq_c.dom1_score]
    cdd_rps = pd.read_csv(FILE_REQS["CDD_RPS_FILES"].format(sample=sample),
                      delim_whitespace=True,
                       names=["qseqid", "sseqid" , "evalue", "pident", "score", "qstart",
                              "qend", "sstart", "send", "length", "slen"])
    cdd_rps["COG"] = cdd_rps.sseqid.apply(lambda x: cdd2cog[x.split("|")[2]])
    cdd_rps["contig"] = cdd_rps.qseqid.apply(lambda x: x.split("_")[0])
    cdd_rps_c = cdd_rps.merge(clustdf, how="inner", on="contig")
    scg_rps = cdd_rps_c[cdd_rps_c.COG.isin(scgs_cogs)]
    hmmvq_best_pog = hmmvq_best.rename(columns={"target":"POG"}).merge(pog_vq,on="POG")
    
    # get number of proteins and contigs
    df3 = pd.DataFrame([
                        (r.id, r.id.split("_")[0], len(r.seq)) for r in \
                        SeqIO.parse(FILE_REQS["PREDICTED_PROT_FAA"].format(sample=sample),
                                    "fasta")
                        ],
                   columns=["protein","contig","protein_length"])
    df3 = df3.merge(clustdf, how="right", on="contig")
    df4 = df3 \
         .merge(hmmvq_best_pog.rename(columns={"query":"protein"}).drop("cluster_gt{c}".format(c=cutoff),axis=1),
          how='left',
          on="protein") \
         .merge(cdd_rps.rename(columns={"qseqid":"protein"}),
                how="left",
                on="protein") \
         .merge(pd.read_csv(FILE_REQS["CONCOCT_INPUT_TABLE"].format(sample=sample),
                            sep="\t").rename(columns={"length":"contig_length"}),
                how="outer",
                on="contig")
    # get bin sizes in bp
    
    
    
    
    # plot it
    fig, axes = plt.subplots(nrows=2, figsize=(16, 10),
                           subplot_kw={'axisbg':'#EEEEEE',
                                       'axisbelow':True})
    df4["is_SCG"] = df4.COG.isin(scgs_cogs)
    df1 = pd.DataFrame(df4.groupby("protein").first().groupby("cluster_gt{c}".format(c=cutoff)).is_SCG.agg({"nr_SCG":np.count_nonzero}))
    df2 = pd.DataFrame(df4[df4["Viral Quotient"] == 1].groupby("protein").first().groupby("cluster_gt{c}".format(c=cutoff)).POG.agg({"nr_POG VQ==1":np.count_nonzero}))
    dfa1 = pd.concat([df1.rename(columns={"is_SCG":"SCG"}), df2], axis=1)
    dfa1.set_index(pd.Int64Index(dfa1.index), inplace=True)
    dfa1.plot(kind="bar", ax=axes[0],
              title="SCG and POG hits per cluster_gt{c} {sample}".format(c=cutoff, sample=sample))
    axes[0].set_ylim(0, 200)
    
    clustdf_len = clustdf \
                .merge(pd.read_csv(FILE_REQS["CONCOCT_INPUT_TABLE"].format(sample=sample),
                            sep="\t").rename(columns={"length":"contig_length"}),
                how="inner",
                on="contig")
    df5 = pd.concat([clustdf_len.groupby("cluster_gt{c}".format(c=cutoff)).contig_length.agg({"nr_bases":np.sum, "nr_contigs":np.count_nonzero}),
                     df4.groupby("cluster_gt{c}".format(c=cutoff)).protein.nunique()], axis=1)
    df5.set_index(pd.Int64Index(df5.index), inplace=True)
    df5.rename(columns={"protein":"nr_proteins"}, inplace=True)
    df5.plot(kind="bar", ax=axes[1], logy=True,
          ylim=(0.9, 1e8),
          title="#contigs, #bases, #proteins per cluster_gt{c} {sample}".format(c=cutoff, sample=sample))
    
    return fig
```

```python
si = StaticInteract(plot_bin_overview,
               cutoff=RadioWidget(PARAMETERS["cutoff"]),
               sample=RadioWidget(PARAMETERS["samples"]))
si
```
